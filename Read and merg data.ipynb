```python
import glob
import os
import pandas as pd
import numpy as np
from datetime import timedelta
```

```python
#Get list of files from train data directory
path = "/mnt/gabers/HDD_data/train_files/"
all_files = glob.glob(os.path.join(path, "*.csv.gz"))
```

```python
#Read the data into one df
df_from_each_file = []
for f in all_files:
    df_from_each_file.append(pd.read_csv(f))
full_train_df   = pd.concat(df_from_each_file, ignore_index=True) 
full_train_df.shape
```

```python
full_train_df["date"] = pd.to_datetime(full_train_df["date"])
full_train_df["date"] = full_train_df["date"].apply(lambda d: d.date())
full_train_df.sort_values(['serial','date'])
```

```python
cat_to_numeric = ["atascttempmaxoplimit", "atascttempminoplimit", "atascttempoverlimit", "atascttempunderlimit"]
# remove 'C'
for column in cat_to_numeric:
    full_train_df[column] = full_train_df[column].str.extract('(^-?[0-9]?[0-9])', expand=False)
    
# convert to numeric
full_train_df[cat_to_numeric] = full_train_df[cat_to_numeric].apply(pd.to_numeric, errors='ignore')
```
